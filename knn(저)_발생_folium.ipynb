# folium으로 정점 찍기

## 빈산소 수괴 발생 정점

```python
df2['어장명'].unique()
```

```python
import folium

m = folium.Map(
    location=[35.0211, 128.6810],
    zoom_start=10
)

tooltip = 'Click'

for i in range(df_저.shape[0]):
    if df_저.iloc[i]["어장명"] == "감포":
        folium.Marker(
            [df_저.iloc[i]['위도'], df_저.iloc[i]['경도']],
            popup=f'<strong>{df_저.iloc[i]["어장명"]}</br>{df_저.iloc[i]["정점2"]}</strong>',
            tooltip=tooltip
        ).add_to(m)

m
```
