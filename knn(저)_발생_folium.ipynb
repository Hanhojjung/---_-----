가막만 발생 데이터 folum

```python
m = folium.Map(
    location=[35.0211, 128.6810],
    zoom_start=10
)

tooltip = 'Click'

for i in range(df2.shape[0]):
    if df2.iloc[i]["어장명"] == "가로림만":
        folium.Marker(
            [df2.iloc[i]['위도'], df2.iloc[i]['경도']],
            popup=f'<strong>{df2.iloc[i]["어장명"]}</br>{df2.iloc[i]["정점2"]}</strong>',
            tooltip=tooltip
        ).add_to(m)

m
```
